```python
import pandas as pd;
import numpy as np;
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Flipkart_Mobiles.csv")
df
```

```python
# 1. Bar Chart: Average Selling Price by Brand
plt.figure(figsize=(10, 6))
avg_selling_price_by_brand = df.groupby('Brand')['Selling Price'].mean().sort_values()
avg_selling_price_by_brand.plot(kind='bar', color='skyblue')
plt.title('Average Selling Price by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Selling Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Pie Chart: Distribution of Ratings
plt.figure(figsize=(8, 8))
rating_distribution = df['Rating'].value_counts()
plt.pie(rating_distribution, labels=rating_distribution.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Rating Distribution')
plt.show()
```

```python
# 3. Histogram: Distribution of Selling Prices
plt.figure(figsize=(10, 6))
sns.histplot(df['Selling Price'], bins=30, kde=True, color='green')
plt.title('Distribution of Selling Prices')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# 4. Timeline Chart: Selling Price over Time
# Assuming 'model' represents chronological order or a proxy for release date
df_sorted = df.sort_values('Model')  # Sorting by model
plt.figure(figsize=(10, 6))
plt.plot(df_sorted['Model'], df_sorted['Selling Price'], marker='o', color='blue')
plt.title('Selling Price Over Models')
plt.xlabel('Model')
plt.ylabel('Selling Price')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Scatter Plot: Selling Price vs. Rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Rating', y='Selling Price', data=df, hue='Brand', palette='viridis',sizes=(100,1000), legend='auto')
plt.title('Selling Price vs. Rating')
plt.xlabel('Rating')
plt.ylabel('Selling Price')
plt.show()
```

```python
#6. Bubble Plot: Rating vs. Selling Price Across Brands
df_scatter = df[['Rating','Selling Price','Brand']]
plt.figure(figsize=[15,8])



sns.scatterplot(x=df_scatter['Rating'], y=df_scatter['Selling Price'],
hue=df_scatter['Brand'], palette=['#F05454','#30475E','#222831'],
size=df_scatter["Rating"], sizes=(100,1000), legend='auto')
plt.title("Rating vs Selling Price Across Different Mobile Brands", size=20, pad=20)
plt.show()
```
