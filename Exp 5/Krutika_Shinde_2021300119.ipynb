```python
# Create Word Cloud

%%R
install.packages("wordcloud")
install.packages("tm")
install.packages("RColorBrewer")

library(wordcloud)
library(tm)
library(RColorBrewer)

# Load dataset
housing <- read.csv('housing.csv')

# Create a corpus from the `median_income` column
corpus <- Corpus(VectorSource(housing$ocean_proximity))

# Preprocess the text data
corpus <- tm_map(corpus, content_transformer(tolower))
corpus <- tm_map(corpus, removePunctuation)
corpus <- tm_map(corpus, removeNumbers)
corpus <- tm_map(corpus, removeWords, stopwords("en"))

# Create a term-document matrix
dtm <- TermDocumentMatrix(corpus)
m <- as.matrix(dtm)
word_freqs <- sort(rowSums(m), decreasing = TRUE)

# Ensure word frequencies are numeric and non-negative
word_freqs <- word_freqs[word_freqs > 0]

# Plot the word cloud
wordcloud(names(word_freqs), word_freqs, min.freq = 1, colors = brewer.pal(8, "Dark2"))
```

**Observation :** The word cloud visualizes the frequency of different ocean_proximity categories in the dataset. Larger and more prominent words represent more frequently occurring proximity categories to the ocean, highlighting the most common locations relative to the ocean.

```python
import plotly.graph_objects as go
import pandas as pd

# Load dataset
housing = pd.read_csv('housing.csv')

# Create 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=housing['longitude'],
    y=housing['latitude'],
    z=housing['median_house_value'],
    mode='markers',
    marker=dict(size=5, color=housing['median_house_value'], colorscale='Viridis', opacity=0.8)
)])
fig.update_layout(title='3D Scatter Plot of Housing Data', scene=dict(
    xaxis_title='Longitude',
    yaxis_title='Latitude',
    zaxis_title='Median House Value'
))

# Show the plot
fig.show()
```

**Observation:** The 3D scatter plot visualizes the relationship between longitude, latitude, and median_house_value. It helps identify spatial patterns and clusters in housing data, showing how house values are distributed geographically.image.png
